# Polynomial Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
icecream_data = pd.read_csv('Ice_cream selling data.csv')
icecream_data.head(5)
```

```python
icecream_data.isnull().sum()
```

```python
icecream_data.rename(columns={
        'Temperature (°C)': 'temperature',
        'Ice Cream Sales (units)': 'sales'
    }, inplace = True)
```

```python
icecream_data.corr()
```

```python
sns.scatterplot(x = icecream_data['temperature'], y = icecream_data['sales'])
```

```python
x = icecream_data[['temperature']]
y = icecream_data['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree = 2)
pf.fit(x)
```

```python
x = pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x.ndim
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
prd = lr.predict(x)
```

```python
sns.scatterplot(x = icecream_data['temperature'], y = icecream_data['sales'], label = 'org')
plt.plot(x, prd, label = 'regLine')
plt.legend()
plt.show()
```
