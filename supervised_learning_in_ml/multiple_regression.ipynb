# Multiple linear regression

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
income_data = pd.read_csv('multiple_linear_regression_dataset.csv')
income_data.head(10)
```

```python
sns.pairplot(income_data)
```

```python
sns.heatmap(data = income_data.corr(), annot = True)
plt.show()
```

```python
x = income_data.iloc[:,:-1]
y = income_data['income']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_prd = lr.predict(x_train)
```

```python
plt.figure(figsize=(6, 6))
sns.regplot(x=y_train, y= y_prd, scatter_kws={'alpha':0.5}, line_kws={'color':'red', 'linestyle':'-'})
plt.title("Actual vs. Predicted Values")
plt.show()
```
