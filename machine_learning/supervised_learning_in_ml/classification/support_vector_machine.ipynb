# Support Vector Machine in Classification

- SVM is one of the most popular algorithm in Supervised Learning algorithms, which is used for Classification as well as Regression problem.

- **Hard Margin:** The algorithm aims to find a hyperplane that perfectly separates the data intom two classes without any misclassification.

- **Soft Margin:** The algorithm allows for some misclassifications to find a hyperplane that generalizes better to unseen data and is more robust to outliers.

                                            **y = mx+c**
                                            **y = mx+c-lambda**
                                            **y = mx+c+lambda**

###### Types of SVMs:

- There are two types of SVMs, each used for different things:

- **Simple SVM:** Typically used for linear regression and classification problems.
- **Kernal SVM:** Has more flexibility for non-linear data because you can add more features to fit a hyperplane instead of a two-dimensional space.

###### Kernel Functions

- Kernal functions play a crucial role in transforming input data into a higher-dimensional space.

- The primary purpose of kernal functions is to allow SVMs to handle non-linearly separable data by implicity mapping the input data into a higher-dimensional feature space where linear separation may be more feasible.

- This transformation is done without explicity calculating the coordinates points in that higher-dimensional space.

```python
from sklearn.model_selection import train_test_split
```

```python
x = dataset.iloc[:, :-1]
y = dataset["placement"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state = 30)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel = "linear")
```

```python
svc.fit(x_train, y_train)
```

```python
print(svc.score(x_train, y_train)*100, svc.score(x_test, y_test)*100)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf = svc)
```
