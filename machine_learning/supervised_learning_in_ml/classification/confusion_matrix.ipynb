# Confusion Matrix

- A confusion matrix is a simple and useful tool for understanding the performance of a classification model, like onenused in Machine learning or statistics

- It helps you evaluate how well your model is doing in categorizing things correctly.

- It is also known as the error matrix.

- The matrix consists of predictions results in a summarized form, which has a number of correct predictions and incorrect predictions.

- **Accuracy =** TP+TN/N

- **Error =** FN+FP/N

- **False Negative:** The model has predicted no, but the actual value was yes, it is also called as **TYPE-II ERROR**

- **False Positive:** The model has predicted Tes, but the actual value was no, it is also called as **TYPE-I ERROR**

#### Precision

**Precision: TP/TP+FP**
- It helps us to measure the ability to classify positive samples in the model

- It tells about how much FP are there. If FP decreases Precision increases. 

#### Recall

**Recall: TP/TP+FN**
- It helps us to measure  how many positive samples were correctly classified by the ML model.

- It tells about how much FN are there. If FN decreases Recall increases.

#### F1-score

- It is the harmonic mean of precision and recall. It takes both false positive and false negative into account. Therefore, it performs well on an imbalanced dataset.

                        F1 Score = 2 * (Precision * Recall) / (Precision + Recall)


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
placed = pd.read_csv(r"/workspaces/Data_Science_Notes/machine_learning/supervised_learning_in_ml/classification/data/placement.csv")
```

```python
placed.head(5)
```

```python
placed = placed.drop(columns = "Unnamed: 0")
```

```python
placed.head(5)
```

```python
x = placed.iloc[:, :-1]
y = placed["placement"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
```

```python
cf = confusion_matrix(y_test, lr.predict(x_test))
```

```python
sns.heatmap(cf, annot = True)
```

```python
precision_score(y_test, lr.predict(x_test))*100
```

```python
recall_score(y_test, lr.predict(x_test))*100
```

```python
f1_score(y_test, lr.predict(x_test))*100
```
