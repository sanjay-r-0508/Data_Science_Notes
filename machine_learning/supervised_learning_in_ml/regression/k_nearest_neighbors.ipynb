# KNN in Regression

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
dataset = pd.read_csv(r"/workspaces/Data_Science_Notes/machine_learning/supervised_learning_in_ml/regression/data/multiple_linear_regression_dataset.csv")
```

```python
dataset
```

```python
dataset.shape
```

```python
x = dataset.iloc[:, :-1]
y = dataset["income"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsRegressor(n_neighbors = 2)
knn.fit(x_train, y_train)
```

```python
print(knn.score(x_train, y_train)*100, knn.score(x_test, y_test)*100)
```

```python
sns.scatterplot(x = '')
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x, y , clf = knn)
```
