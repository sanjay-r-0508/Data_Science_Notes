# Regression Analysis

Regression analysis is a powerful statistical method used in machine learning and statistics to examine the relationship between one dependent variable (the target, often denoted as \(y\)) and one or more independent variables (the features, often denoted as \(X\)). 

The primary goal of regression analysis is to understand how the value of the dependent variable changes when any one of the independent variables is varied. 

##### The Core Goal: Prediction and Understanding
Regression analysis serves two main purposes:

1. Predicting Outcomes: To forecast future values of the dependent variable based on known values of the independent variables. (e.g., predicting the price of a house based on its square footage and location).

2. Understanding Relationships: To determine which independent variables are significantly related to the dependent variable and the strength and direction of that relationship (e.g., confirming that, statistically, an extra bedroom increases a house price by a certain average amount). 

##### Types of Regression:

1. Linear Regression
2. Multi-Linear Regression
3. Lasso Regression
4. Ridge Regression
5. Polynomial Regression
6. Random Forest Regression
7. Support Vector Machine
8. K-Nearest Neighbour



# 1. Simple Linear Regression

Simple Linear is a type of Regression algorithm that models the relationship between a dependent variable and a single independent variable.

Linear Equation:  
                                    y = mx+c

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
package = pd.read_csv('placement.csv')
package.head(10)
```

```python
sns.scatterplot(x='cgpa', y='package', data=package)
plt.title('CGPA vs Package')
plt.show()
```

```python
package.isnull().sum()
```

```python
x = package[['cgpa']]
y = package['package']
```

```python
x.ndim
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
m = lr.coef_
```

```python
c= lr.intercept_
```

```python
lr.score(x_test, y_test)*100
```

```python
lr.predict([[9.5]])
```

```python
m*9.5+c
```

```python
y_pred = lr.predict(x)
```

```python
sns.scatterplot(x='cgpa', y='package', data=package, label='Data Points')
plt.plot(package['cgpa'], y_pred, color='red', label='Regression Line')
plt.legend()
plt.title('CGPA vs Package')
plt.show()
```
